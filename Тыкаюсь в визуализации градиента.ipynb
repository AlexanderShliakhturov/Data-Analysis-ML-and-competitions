```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#  ,      
def f(x, y):
    return x**2 + y**2  + np.sin(x)

#      
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

#       
plt.contour(X, Y, Z, levels=40)

#      
start_point = np.array([4.0, 4.0], dtype=np.float64)

#    
learning_rate = 0.6

#       
num_steps = 2
trajectory_x = [start_point[0]]
trajectory_y = [start_point[1]]
current_point = start_point.copy()
for step in range(num_steps):
    gradient = np.array([2*current_point[0] + np.cos(current_point[0]),
                         2 * current_point[1]], dtype=np.float64)  #   x^2 + y^2
    current_point -= learning_rate * gradient
    trajectory_x.append(current_point[0])
    trajectory_y.append(current_point[1])

#   
plt.plot(trajectory_x, trajectory_y, marker='.', color='red', linestyle='-')

#     
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot with Gradient Descent Trajectory')
plt.grid(True)
plt.colorbar(label='Function Value')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

#  ,      
def f(x, y):
    return x**2 + y**2 + np.sin(x)

#       
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

#       
plt.contour(X, Y, Z, levels=40)

#      
start_point = np.array([4.0, 4.0], dtype=np.float64)

#    
learning_rate = 0.1

#           
num_steps = 20
trajectory_x = [start_point[0]]
trajectory_y = [start_point[1]]
current_point = start_point.copy()
for step in range(num_steps):
    gradient = np.array([2*current_point[0] + np.cos(current_point[0]), 2 * current_point[1]], dtype=np.float64)  #   x^2 + y^2
    current_point -= learning_rate * gradient
    trajectory_x.append(current_point[0])
    trajectory_y.append(current_point[1])

#     
plt.plot(trajectory_x, trajectory_y, marker='.', color='red', linestyle='-', label='Gradient Descent')

#         
momentum = 0.9
trajectory_x_momentum = [start_point[0]]
trajectory_y_momentum = [start_point[1]]
current_point_momentum = start_point.copy()
velocity = np.zeros_like(current_point_momentum)
for step in range(num_steps):
    gradient = np.array([2*current_point_momentum[0] + np.cos(current_point_momentum[0]), 2 * current_point_momentum[1]], dtype=np.float64)  #   x^2 + y^2
    velocity = momentum * velocity - learning_rate * gradient
    current_point_momentum += velocity
    trajectory_x_momentum.append(current_point_momentum[0])
    trajectory_y_momentum.append(current_point_momentum[1])

#      
plt.plot(trajectory_x_momentum, trajectory_y_momentum, marker='.', color='blue', linestyle='-', label='Momentum Descent')

#     
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot with Gradient Descent Trajectories')
plt.grid(True)
plt.colorbar(label='Function Value')
plt.legend()
plt.show()
```
