```python
# Определяем функцию, для которой будем рисовать линии уровня
def f(x, y):
    return x**2 + y**2  + np.sin(x)

# Генерируем данные для построения линий уровня
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Создаем график и добавляем линии уровня функции
plt.contour(X, Y, Z, levels=40)

# Задаем начальную точку для градиентного спуска
start_point = np.array([4.0, 4.0], dtype=np.float64)

# Определяем шаг градиентного спуска
learning_rate = 0.6

# Выполняем градиентный спуск и сохраняем координаты шагов
num_steps = 2
trajectory_x = [start_point[0]]
trajectory_y = [start_point[1]]
current_point = start_point.copy()
for step in range(num_steps):
    gradient = np.array([2*current_point[0] + np.cos(current_point[0]),
                         2 * current_point[1]], dtype=np.float64)  # Градиент функции x^2 + y^2
    current_point -= learning_rate * gradient
    trajectory_x.append(current_point[0])
    trajectory_y.append(current_point[1])

# Отображаем траекторию спуска
plt.plot(trajectory_x, trajectory_y, marker='.', color='red', linestyle='-')

# Добавляем подписи и отображаем график
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot with Gradient Descent Trajectory')
plt.grid(True)
plt.colorbar(label='Function Value')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию, для которой будем рисовать линии уровня
def f(x, y):
    return x**2 + y**2 + np.sin(x)

# Генерируем данные для построения линий уровня функции
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Создаем график и добавляем линии уровня функции
plt.contour(X, Y, Z, levels=40)

# Задаем начальную точку для градиентного спуска
start_point = np.array([4.0, 4.0], dtype=np.float64)

# Определяем шаг градиентного спуска
learning_rate = 0.1

# Выполняем градиентный спуск и сохраняем координаты шагов для классического градиентного спуска
num_steps = 20
trajectory_x = [start_point[0]]
trajectory_y = [start_point[1]]
current_point = start_point.copy()
for step in range(num_steps):
    gradient = np.array([2*current_point[0] + np.cos(current_point[0]), 2 * current_point[1]], dtype=np.float64)  # Градиент функции x^2 + y^2
    current_point -= learning_rate * gradient
    trajectory_x.append(current_point[0])
    trajectory_y.append(current_point[1])

# Отображаем траекторию классического градиентного спуска
plt.plot(trajectory_x, trajectory_y, marker='.', color='red', linestyle='-', label='Gradient Descent')

# Выполняем градиентный спуск с моментом и сохраняем координаты шагов
momentum = 0.9
trajectory_x_momentum = [start_point[0]]
trajectory_y_momentum = [start_point[1]]
current_point_momentum = start_point.copy()
velocity = np.zeros_like(current_point_momentum)
for step in range(num_steps):
    gradient = np.array([2*current_point_momentum[0] + np.cos(current_point_momentum[0]), 2 * current_point_momentum[1]], dtype=np.float64)  # Градиент функции x^2 + y^2
    velocity = momentum * velocity - learning_rate * gradient
    current_point_momentum += velocity
    trajectory_x_momentum.append(current_point_momentum[0])
    trajectory_y_momentum.append(current_point_momentum[1])

# Отображаем траекторию градиентного спуска с моментом
plt.plot(trajectory_x_momentum, trajectory_y_momentum, marker='.', color='blue', linestyle='-', label='Momentum Descent')

# Добавляем подписи и отображаем график
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot with Gradient Descent Trajectories')
plt.grid(True)
plt.colorbar(label='Function Value')
plt.legend()
plt.show()
```
